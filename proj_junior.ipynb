## 1 - Carregar os arquivos e cadastro, investimento1 e investimento2 e mostrar a quantidade de linhas de cada um, além das informações desses dataframes

```python
import pandas as pd
import datetime as dt

cad = pd.read_csv('cadastro.csv')
invest1 = pd.read_csv('investimento_parte1.csv')
invest2 = pd.read_csv('investimento_parte2.csv')
print(f'Quantidade de linhas dos arquivos csv\ncad: {cad.shape[0]}  invest1:{invest1.shape[0]} invest2: {invest2.shape[0]}')

display(cad.head(3))
display(invest1.head(3))
display(invest2.head(3))
```

```python
# Preenchendo os valores faltantes com a média
dataset_mean = dataset.copy()
media_y = dataset.Ativo01.mean()
print(media_y)
dataset_mean['Ativo01'] = dataset_mean['Ativo01'].fillna(0)
dataset_mean['Ativo02'] = dataset_mean['Ativo02'].fillna(0)
dataset_mean['Ativo03'] = dataset_mean['Ativo03'].fillna(0)
dataset_mean['Ativo04'] = dataset_mean['Ativo04'].fillna(0)
dataset_mean['Ativo05'] = dataset_mean['Ativo05'].fillna(0)
dataset_mean['Ativo06'] = dataset_mean['Ativo06'].fillna(0)
dataset_mean['Nota1'] = dataset_mean['Nota1'].fillna(dataset.Nota1.mean())
dataset_mean['Nota2'] = dataset_mean['Nota2'].fillna(dataset.Nota2.mean())
dataset_mean.isnull().sum()
```

## 7 - Manipule os dados a seguir:
- Criar coluna "Total" somando o valor de todos os ativos por investidor
- Criar coluna "Media" com a média das notas 1 e 2
- Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)
- Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)
- Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar o map)
- Transformar os dados da coluna Media em inteiro
- Criar coluna "NPS" onde o investidor é promotor se a média das notas é maior igual a 80, neutro se a média estiver entre 50 e 80, e detrator se for abaixo de 50


```python
dataset_mani = dataset_mean.copy()
dataset_mani.loc[:, 'Total'] = dataset_mani['Ativo01'] + dataset_mani['Ativo02'] + dataset_mani['Ativo03'] + dataset_mani['Ativo04'] + dataset_mani['Ativo05'] + dataset_mani['Ativo06'] + dataset_mani['Ativo07']
dataset_mani.loc[:, 'Media'] = (dataset_mani['Nota1'] + dataset_mani['Nota2'])/2

#mudando o nome da coluna salario para renda
dataset_mani = dataset_mani.rename(columns = {'Salario': 'Renda'})
dataset_mani['Renda'] = dataset_mani['Renda'].str.replace('$', '').astype(float)
dataset_mani.head(3)
```

```python
dataset_mani['Sexo'] = dataset_mani['Sexo'].replace('Fem', 'Feminino')
dataset_mani['Sexo'] = dataset_mani['Sexo'].replace('Masc', 'Masculino')
dataset_mani.head(3)
```

```python
# Utilizando o apply a
dataset_mani['NPS'] = dataset_mani.Media.apply(lambda x: 'Promotor' if x > 80 else('Normal' if x > 50 else 'Detrator'))
dataset_mani.head(3)
```
